# Class

````python
class Parent:
    def __init__(self, text):
        self.text = text

class Child(Parent):
    def __init__(self, text):
        super().__init__(text)
        # super().__init__(text)

variavel = Child("hello")
print(variavel.text)  # hello
````

### Explicação linha por linha do código acima
1. `class Parent:`<br>Define uma classe chamada Parent ("Pai").
2. `def __init__(self, text):`<br>Cria o método construtor da classe Parent, que recebe o parâmetro `text`.
3. `self.text = text`<br>Salva o valor recebido em `text` como um atributo do objeto.
4. `class Child(Parent):`<br>Define uma classe chamada Child ("Filho"), que herda de Parent.
5. `def __init__(self, text):`<br>Cria o construtor da classe Child, também recebendo o parâmetro `text`.
6. `super().__init__(text)`<br>Chama o construtor da classe Parent para inicializar o atributo `text` herdado.
7. `# super().__init__(text)`<br>Comentário, não é executado.
8. `variavel = Child("hello")`<br>Cria um objeto da classe Child, passando "hello" como argumento.
9. `print(variavel.text)  # hello`<br>Imprime o valor do atributo `text` do objeto, que será "hello".

#### Exemplo simples de herança em Python
```python
class Animal:
    def falar(self):
        print("Algum som...")

class Cachorro(Animal):
    pass

bobby = Cachorro()
bobby.falar()  # Saída: Algum som...
```
No exemplo acima, a classe `Cachorro` herda o método `falar` da classe `Animal`.

````python
class Parent:
    def __init__(self, text):
        self.text = text

class Child(Parent):
    def __init__(self, text):
        super().__init__(text)


Roberto = Child("bonito")
print(Roberto.text)  # bonito
````

### Which of the following data structures are mutable in Python?
- **tuple** ❌ (immutable)
- **string** ❌ (immutable)
- **set** ✅ (mutable)
- **dictionary** ✅ (mutable)
- **list** ✅ (mutable)

**Mutable** means you can change (add, remove, or modify) the elements after creation.

````python
# Operador lógico OR em Python
a = True
b = False
print(a or b)  # Saída: True
print(False or False)  # Saída: False
````
